```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
drift = np.linspace(25, 625, 13)
error = np.array([ 1177.68, 104.81, 60.94, 47.97, 38.66, 36.39, 30.05, 26.51, 25.41, 24.16, 20.36, 22.7, 30.57 ])
```

```python
fig, ax = plt.subplots( figsize = (4, 3), layout = 'tight' )
ax.plot(drift, error, marker = 'o', linestyle = '-')
```

```python
drift = np.linspace(75, 625, 12)
error = np.array([ 104.81, 60.94, 47.97, 38.66, 36.39, 30.05, 26.51, 25.41, 24.16, 20.36, 22.7, 30.57 ])
```

```python
fig, ax = plt.subplots( figsize = (8, 6), layout = 'tight' )

ax.plot(drift, error, marker = 'o', linestyle = '-')
ax.set_xlabel('Drift Length [cm]', fontsize = 22)
ax.set_ylabel('Relative Drift Length Error [%]', fontsize = 22)
ax.tick_params(axis = 'both', which = 'major', labelsize = 22)
ax.set_xlim([0, 650])
ax.grid(True)
```

```python
old_drift = np.array([50, 130, 250, 410, 610])
old_error = np.array([27.01, 18.83, 15.23, 12.91, 10.51])
```

```python
fig, ax = plt.subplots( figsize = (10, 5), layout = 'tight' )

ax.plot(old_drift, old_error, marker = 'o', linestyle = '-', label = 'GEANT4+diffusion simulation')
ax.plot(drift, error, marker = 'o', linestyle = '-', label = 'Full simulation')
ax.set_xlabel('Drift Length [cm]', fontsize = 22)
ax.set_ylabel('Relative Drift Length Error [%]', fontsize = 22)
ax.set_title(r'Pixel: 5mm, Noise level: 50e$^-$, Tile: 10cm; first attempt', fontsize = 22)
ax.tick_params(axis = 'both', which = 'major', labelsize = 22)
ax.set_xlim([0, 650])
ax.grid(True)

ax.legend(fontsize = 22)
fig.savefig('DriftDistance.pdf')
fig.savefig('DriftDistance.png')
```
