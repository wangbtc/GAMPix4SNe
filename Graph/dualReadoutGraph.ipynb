```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gaussian parameters
mu = 0
sigma = 1
```

```python
# X values
x = np.linspace(-5, 5, 1000)
```

```python
# Gaussian PDF
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

# Plot the curve
ax.plot(x, y, color = 'gray', linewidth = 4, alpha = 0.7, label = 'Gaussian')

# Fill from -2σ to +1.5σ
mask1 = (x >= mu - 2 * sigma) & (x <= mu + 1.5 * sigma)
ax.fill_between(x[mask1], y[mask1], color = 'darkorange', alpha = 0.5, label = '-1.5σ to +1.5σ')

# Fill from +1.5σ to +5σ
mask2 = (x >= mu + 1.5 * sigma) & (x <= mu + 5 * sigma)
ax.fill_between(x[mask2], y[mask2], color = 'yellowgreen', alpha = 0.5, label = '+1.5σ to +4.5σ')

# Optional: left side from -5σ to -2σ
mask3 = (x >= mu - 5 * sigma) & (x <= mu - 2 * sigma)
ax.fill_between(x[mask3], y[mask3], color = 'lightblue', alpha = 0.7)

ax.axhline( y = 0, xmin = 0.05, xmax = 0.95, color = 'gray', linestyle = '--', linewidth = 4, alpha = 0.7)

# Sigma guide lines
# for k in [-1.5, 1.5, 4.5]:
#     ax.axvline(mu + k * sigma, color='gray', linestyle='--', alpha=0.6)
ax.set_xlim([-4.5, 4.5])
ax.axis('off')
# ax.set_title('Gaussian with Colored Sigma Regions')
# ax.set_xlabel('x')
# ax.set_ylabel('Density')
# ax.legend()
# ax.grid(alpha = 0.3)
```
